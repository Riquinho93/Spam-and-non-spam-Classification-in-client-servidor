```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# ML Packages For Vectorization of Text For Feature Extraction
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Read the data
df = pd.read_csv("spam_or_not_spam.csv")
```

```python
# Print the 5th first data
df.head()
```

```python
# total data
df.size
```

```python
# Checking for null values
df.isnull().isnull().sum()
```

```python
df_data = df[["email","label"]]
```

```python
df_data.columns
```

```python
df_x = df_data['email']
df_y = df_data['label']
```

```python
# Convert a collection of text documents to a matrix of token counts
cv = CountVectorizer()
ex = cv.fit_transform(["Great song but check this out","What is this song?"])
```

```python
ex.toarray()
```

```python
# See the vectorization
cv.get_feature_names()
```

```python
corpus = df_x

corpus
```

```python
cv = CountVectorizer()
cv
```

```python
X = cv.fit_transform(corpus.astype('U').values) # Fit the Data
```

```python
X.toarray()
```

```python
cv.get_feature_names()
```

```python
# Import for train and test the model
from sklearn.model_selection import train_test_split
```

```python
# Implementation of the model
X_train, X_test, y_train, y_test = train_test_split(X, df_y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
# Implementation of Naive Bayes
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(X_train,y_train)
clf.score(X_test,y_test)
```

```python
# Accuracy of our Model
print("Accuracy of Model",clf.score(X_test,y_test)*100,"%")
```

```python
## Predicting with our model
clf.predict(X_test)
```

```python
# Sample Prediciton
comment = ["Check this out"]
vect = cv.transform(comment).toarray()
```

```python
clf.predict(vect)
```
